```python
list_class =[]
hits=[]
with open('Resfams4_29_hits.tsv', 'r') as f:
    for line in f:
        if line.startswith('Query:       '):
            print(line.split('|')[2])
            list_class.append(line.split('|')[2])
            #print( re.split( '      |\n|  |    |  |   |    |  |   |    '  , next(f).strip(','))[-7:])
            for string in re.split( '      |\n|  |    |  |   |    |  |   |    '  , next(f).strip(','))[-5:-2]:
                if string != '' and re.match("^[a-zA-Z]+.*", string[0]):
                    print(string)
                    hits.append(string)
                    
    f.close()
#print(len(hits))    
#print(len(list_class))
#print(hits)
```

```python
unique_classes= list(set(list_class))

print(list(set(list_class)))
```
